```python
import os
import pandas as pd
import numpy as np

def load_and_combine_datasets(directory):
    combined_df = pd.DataFrame()
    for filename in os.listdir(directory):
        if filename.endswith(".csv") and 'GSE' in filename:
            filepath = os.path.join(directory, filename)
            # Read the dataset with the correct header interpretation
            df = pd.read_csv(filepath, index_col=0).T  # Transpose so samples are rows
            # Filter the dataset to include only the columns of interest
            combined_df = pd.concat([combined_df, df], ignore_index=True)
    
    return combined_df

directory_path = '.'

combined_data = load_and_combine_datasets(directory_path)
#train_and_evaluate_model(combined_data)
combined_data
```
